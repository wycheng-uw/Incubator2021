```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc4
```

```python
xr.show_versions()
```

```python
capeurl = "http://iridl.ldeo.columbia.edu/SOURCES/.Models/.SubX/.EMC/.GEFS/.hindcast/.cape/dods"
prurl   = "http://iridl.ldeo.columbia.edu/SOURCES/.Models/.SubX/.EMC/.GEFS/.hindcast/.pr/dods"
```

```python
remote_capedata = xr.open_dataset(capeurl,chunks={'S':'auto', 'M':'auto', 'L':'auto', 'Y':'auto','X':'auto'})
remote_prdata   = xr.open_dataset(prurl,  chunks={'S':'auto', 'M':'auto', 'L':'auto', 'Y':'auto','X':'auto'})
```

```python
display(remote_capedata)
display(remote_prdata)
```

```python
capedata = remote_capedata['cape']
prdata = remote_prdata['pr']
```

```python
capedata_mean = capedata.mean(dim={'S','M','L'})
prdata_mean   = prdata.mean(dim={'S','M','L'})
```
